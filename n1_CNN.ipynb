```python
# Import Libraries
import numpy as np
import tensorflow as tf
import sys as sy
import struct as st
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Input data
fI = open('/home2/bamboo/NeuralNetForPredict/Images/n1/ewpLH-ImgRes_0.dig', 'rb' ) 
nxI = fI.read(4)
nxI = st.unpack("i", nxI)[0]
nyI = fI.read(4)
nyI = st.unpack("i", nyI)[0]
img_target = []
for _ in range(nxI*nyI):
    data = fI.read(8)
    data = st.unpack("d", data)[0]
    img_target.append(data)
fI.close()
target = np.array(img_target).reshape(nyI, nxI)
fIL = open('/home2/bamboo/NeuralNetForPredict/Images/n1/ewpLH-LImgRes_0.dig', 'rb')
nxIL = fIL.read(4)
nxIL = st.unpack("i", nxIL)[0]
nyIL = fIL.read(4)
nyIL = st.unpack("i", nyIL)[0]
img_low = []
for _ in range(nxIL*nyIL):
    data = fIL.read(8)
    data = st.unpack("d", data)[0]
    img_low.append(data)
fIL.close()
low = np.array(img_low).reshape(nyIL, nxIL)
```

```python
# Make dataset
numberOfTargets = 2
width, height = nxI, nyI
width_block, height_block = 6, 6
size_block = width_block * height_block
halfOfWidthBlock = int(width_block * 0.5)
width_low, height_low = width_block, 1
size_low = width_low * height_low
width_target, height_target = width_block, height_block - height_low
size_target = width_target * height_target
arr_data = []
for y in range(height - height_block + 1):
    for x in range(0, width - width_block + 1, numberOfTargets):
        target_block = target[y:y+height_target, x:x+width_target].reshape(size_target, 1)
        arr_data.extend(target_block)
        low_block = low[y+height_target, x:x+width_low].reshape(size_low, 1)
        arr_data.extend(low_block)
train_data = np.array(arr_data)
train_x = train_data.reshape(int(train_data.size/size_block), size_block)

arr_data = []
for y in range(height_block - 1, height, 1):
    for x in range(halfOfWidthBlock - 1, width - (halfOfWidthBlock - 1), numberOfTargets):
        label = target[y, x:x+numberOfTargets]
        arr_data.append(label)
label = np.array(arr_data)
train_label = np.reshape(label, (len(train_x), numberOfTargets))
```
